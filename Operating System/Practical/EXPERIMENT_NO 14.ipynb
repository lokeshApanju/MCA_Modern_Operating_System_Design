```python
# Program to monitor file access pattern and suggest cache strategy

def suggest_cache_strategy(access_list):
    total_access = len(access_list)
    unique_files = len(set(access_list))

    print("File Access Pattern:", access_list)
    print("Total Accesses:", total_access)
    print("Unique Files:", unique_files)

    if unique_files < total_access / 2:
        print("Suggested Cache Strategy: LRU (Least Recently Used)")
    else:
        print("Suggested Cache Strategy: FIFO (First In First Out)")


# Example file access pattern
files = ['A', 'B', 'A', 'C', 'A', 'B', 'D', 'A']

suggest_cache_strategy(files)
```
