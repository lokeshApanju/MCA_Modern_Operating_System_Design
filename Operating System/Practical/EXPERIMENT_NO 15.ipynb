```python
# Simulating Container as a Scheduler (FCFS)
import time

# List of containers (tasks)
containers = [
    ("Container-1", 2),
    ("Container-2", 1),
    ("Container-3", 3)
]

print("Container Scheduler Started\n")

# FCFS Scheduling
for container, cpu_time in containers:
    print(f"Running {container}")
    time.sleep(cpu_time)   # simulate CPU execution
    print(f"{container} completed\n")

print("All containers executed successfully")
```
